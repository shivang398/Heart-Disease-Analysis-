```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df =pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

Check for missing values 

```python
df.isnull().sum()
```

Distributive plot for the person having the disease in the particlar age 

```python
sns.displot(df['Age'])
```

Histogram for the blood pressure of the person having Heart disease 

```python
sns.histplot(df['RestingBP'],color= 'black')
```

Distributive Plot for Cholestrol

```python
sns.displot(df['Cholesterol'],color= 'red')
```

Piechart for the gender 

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%.2f')
```

PieChart for RestingECG

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct='%.2f')
```

Pie chart for chest pain

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%.2f')
```

Violin Plot for the Age and Heart Disease 

```python
sns.violinplot(y=df['Age'],x=df['HeartDisease'])
```

Joint Plot between Age and Resting Bp

```python
sns.jointplot(x='Age', y='RestingBP', data=df)
```

Pair Plot

```python
sns.pairplot(df)
```
